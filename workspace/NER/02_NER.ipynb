```python
# Import libraries
import pandas as pd
import spacy
```

```python
#  Load dataset
df = pd.read_csv("../Data/raw/Students_Performance_dataset.csv")
```

```python
# Focus on relevant columns
skills_col = "What are the skills do you have ?"
interests_col = "What is you interested area?"
```

```python
skills_data = df[skills_col].dropna().tolist()
interests_data = df[interests_col].dropna().tolist()
```

```python

print("Sample skills:", skills_data[:5])
print("Sample interests:", interests_data[:5])
```

```python
# Load spaCy model
nlp = spacy.load("en_core_web_sm")
```

```python
# Extract entities from skills & interests
def extract_entities(texts):
    entities = []
    for text in texts:
        doc = nlp(str(text))
        entities.extend([ent.text for ent in doc.ents])
    return entities

skills_entities = extract_entities(skills_data)
interests_entities = extract_entities(interests_data)
```

```python
print("Extracted skills entities:", skills_entities[:10])
print("Extracted interests entities:", interests_entities[:10])
```

```python
# Build frequency tables
skills_freq = pd.Series(skills_entities).value_counts()
interests_freq = pd.Series(interests_entities).value_counts()

print("\nTop Skills:\n", skills_freq.head(10))
print("\nTop Interests:\n", interests_freq.head(10))
```
