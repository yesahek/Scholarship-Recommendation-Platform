```python
#Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv("Data/raw/Students_Performance_dataset.csv")

# Step 3: Quick overview
print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nFirst 5 rows:")
df.head()
```

```python
#Check missing values
print("\nMissing values per column:")
print(df.isnull().sum())
```

```python
#Scholarship status distribution
plt.figure(figsize=(6,4))
sns.countplot(x="Do you have meritorious scholarship ?", data=df, palette="viridis")
plt.title("Scholarship Status Distribution")
plt.show()
```

```python
#CGPA distribution
plt.figure(figsize=(6,4))
sns.histplot(df["What is your current CGPA?"], bins=20, kde=True, color="blue")
plt.title("CGPA Distribution")
plt.xlabel("CGPA")
plt.show()
```

```python
#Family income vs Scholarship
plt.figure(figsize=(8,6))
sns.boxplot(x="Do you have meritorious scholarship ?", y="What is your monthly family income?", data=df)
plt.title("Family Income vs Scholarship Status")
plt.show()
```

```python
# Correlation heatmap (numerical features only)
plt.figure(figsize=(10,8))
sns.heatmap(df.select_dtypes(include="number").corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
